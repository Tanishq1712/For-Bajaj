```python
def is_valid_indian_number(number):
    # Check if the number starts with 6, 7, 8, or 9 and has 10 digits.
    return str(number).startswith(('6', '7', '8', '9')) and len(str(number)) == 10

# Apply the function to the 'phoneNumber' column and create a new column 'isValidMobile'
data_csv['isValidMobile'] = data_csv['phoneNumber'].apply(is_valid_indian_number)

# Count the number of rows where 'isValidMobile' is True
valid_phone_count = data_csv['isValidMobile'].sum()

# Print the result
print(valid_phone_count)
```

```python


import pandas as pd
# Fill missing birthdates with a placeholder
data_csv['birthDate'] = data_csv['birthDate'].fillna('Unknown')

# Convert birthdates to datetime objects, handling errors
def convert_to_datetime(date_str):
    try:
        return pd.to_datetime(date_str)
    except:
        return pd.NaT

data_csv['birthDate'] = data_csv['birthDate'].apply(convert_to_datetime)

# Drop rows with invalid birthdates
data_csv.dropna(subset=['birthDate'], inplace=True)
```

```python


from datetime import datetime, date

def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))


data_csv['age'] = data_csv['birthDate'].apply(calculate_age)

# Categorize into age groups
def age_group(age):
    if age < 13:
        return 'Child'
    elif age < 20:
        return 'Teen'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

data_csv['ageGroup'] = data_csv['age'].apply(age_group)

# Count of Adults
adult_count = data_csv[data_csv['ageGroup'] == 'Adult']['ageGroup'].count()
print(adult_count)
```

```python
# Calculate percentage of females
female_percentage = (data_csv['gender'] == 'F').sum() / len(data_csv) * 100

# Print the result
print(female_percentage)
```
